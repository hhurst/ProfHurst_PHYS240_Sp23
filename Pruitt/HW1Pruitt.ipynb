```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta0 = np.linspace(0, 2*np.pi, 20)
x = lambda ang : 2*np.cos(ang) + np.cos(2*ang)
y = lambda ang : 2*np.sin(ang) - np.sin(2*ang)

# printout of x and y values at each theta value
for one, two, three in zip(x(theta0), y(theta0), theta0):
    print(f'theta: {three:.2f} - x: {one:.2f} y: {two:.2f}')
```

```python
plt.figure()
plt.plot(x(theta0), y(theta0), 'k.')
plt.title('Deltoid Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def polarToCart(r, theta):
    
    # function that takes 2 inputs to change polar coordinates to cartesian
    
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return(x, y)
```

```python
theta1 = np.linspace(0, 10*np.pi, 1000)
r1 = theta1**2
plt.figure()
plt.title('Galilean Spiral')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(*polarToCart(r1, theta1), 'k')
```

```python
theta2 = np.linspace(0, 24*np.pi, 10000)

r2 = np.exp(np.cos(theta2)) - 2*np.cos(4*theta2) + np.sin(theta2/12)**5

plt.figure()
plt.plot(*polarToCart(r2, theta2), 'k')
plt.title('Feys Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
