```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1,2,3], [0,4,5], [0,0,6]])
B = np.array([[1,0,0], [0,2,0], [0,0,3]])
C = np.array([[1,2,3], [4,5,6], [7,8,3]])
D = np.array([[1,1/2,1/3], [1/4,1/5,1/6], [1/7,1/8,1/9]])

invA = np.linalg.inv(A)
invB = np.linalg.inv(B)
invC = np.linalg.inv(C)
invD = np.linalg.inv(D)
```

```python
# np.allclose(A,B) checks if A and B are elementwise the same within tolerance 
print(np.allclose(np.dot(A, invA), np.eye(3)))
print(np.allclose(np.dot(B, invB), np.eye(3)))
print(np.allclose(np.dot(C, invC), np.eye(3)))
print(np.allclose(np.dot(D, invD), np.eye(3)))
```

```python
eigPairA = np.linalg.eig(A)
eigPairB = np.linalg.eig(B)
eigPairC = np.linalg.eig(C)
eigPairD = np.linalg.eig(D)

print(f'\nEigenvalues of A: {eigPairA[0]}, \n\nEigenvectors of A: {eigPairA[1][0]},\n{eigPairA[1][1]}\n{eigPairA[1][2]}')
print(f'--------------------------------------------------------------------------------------------------------------')
print(f'\nEigenvalues of B: {eigPairB[0]}, \n\nEigenvectors of B: {eigPairB[1][0]},\n{eigPairB[1][1]}\n{eigPairB[1][2]}')
print(f'--------------------------------------------------------------------------------------------------------------')
print(f'\nEigenvalues of C: {eigPairC[0]}, \n\nEigenvectors of C: {eigPairC[1][0]},\n{eigPairC[1][1]}\n{eigPairC[1][2]}')
print(f'--------------------------------------------------------------------------------------------------------------')
print(f'\nEigenvalues of D: {eigPairD[0]}, \n\nEigenvectors of D: {eigPairD[1][0]},\n{eigPairD[1][1]}\n{eigPairD[1][2]}')
```

```python
def sig_figs(x, n):
    # x is the number to be converted to appropriate sig figs
    # n is the appropriate number of sig figs
    
    x = str(x).split('.')
    left = x[0]
    right = x[1]
    return(left + '.' + right[:n - len(left)])
    
```

```python
R = 1.097e-2 # nm^-1

lamb = lambda m, n:  1/(R*((1/m**2) - (1/n**2)))
waveList = []
for m in range(3):
    for n in range(5+m+1):
        if n+1>m+1:
            wave = lamb(m+1, n+1)
            print(f'\nm = {m+1}, n = {n+1} : {sig_figs(wave, 4)} nm')            
```

```python
f1 = lambda x: np.exp(-x/4)*np.sin(x)
x1 = np.linspace(0, 20, 100)

plt.figure()
plt.plot(x1, f1(x1), 'k')
plt.title('$f(x) = e^{-x/4}sin(x)$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

```python
f1 = lambda x: np.exp(-x/4)*np.sin(x)
f2 = lambda x: np.exp(-x/4)
x1 = np.linspace(0, 20, 1000)
x2 = np.linspace(0, 20, 13)

plt.figure()
plt.plot(x1, f1(x1), 'r')
plt.plot(x2, f1(x2), 'ro')
plt.plot(x1, f2(x1), 'b--')
plt.title('$f(x) = e^{-x/4}sin(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

```python
f3 = lambda x: np.exp(-x)*(np.sin(x)**2)
x1 = np.linspace(0, 19, 200)

plt.figure()
plt.semilogy(x1, f3(x1), 'r+')
plt.title('$f(x) = e^{-x}sin^{2}(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.xlim([0, 20])
# plt.ylim([10**-10, 0])
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
r = np.cos(6*theta+np.pi)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r, 'b')
ax.set_rmax(1)
ax.set_rticks([0.5, 1.0])  # Less radial ticks
ax.set_rlabel_position(22.5)  # Move radial labels away from plotted line
ax.set_rlim([0,1.05])
plt.show()
```
